```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
print(x.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg1=LinearRegression()
reg1.fit(x,y)
```

```python
reg1.score(x,y)
```

```python
reg1.coef_ #m
```

```python
reg1.intercept_ #c
```

```python
reg1.predict([[1504.75]])
```

```python
y_pred=reg1.predict(x)
plt.scatter(x,y,color='b')
plt.plot(x,y_pred,color='r')

plt.show()

```
